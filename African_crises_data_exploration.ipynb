```python
df_mean = df[['exch_usd','country', 'banking_crisis']].groupby(['country']).max()
df_mean
```

```python
df_crisis = df[['exch_usd','country', 'banking_crisis','year']][(df['banking_crisis']=='crisis') & (df['year']>=1900)].groupby(['country','banking_crisis']).agg('max')
df_crisis
```

```python
df_crisis.info()
```

```python
df_crisis = df_crisis.reset_index()
df_crisis
```

```python
df_crisis = df_crisis.pivot(index='country',columns='banking_crisis', values=['exch_usd','year'])
df_crisis
```

```python
pivoted_dict = df_crisis.to_dict()
pivoted_dict
```

```python
countries = ['south africa', 'nigeria']
exch_usd = pivoted_dict[('exch_usd', 'crisis')]
year = pivoted_dict[('year', 'crisis')]
result_dict = {}

for (k,v),(k1,v1) in zip(exch_usd.items(), year.items()):
    if k == k1:
        result_dict.update({k :f"exch_usd {v} in crisis year {v1}"})
print(result_dict)
            

#list_returned = [f"{k} had a us exchange rate of  for k,v in pivoted_dict.items() if k.lower() in counties]
```
